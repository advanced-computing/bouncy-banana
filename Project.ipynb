```python
# Library Jargon
import requests
import pandas as pd
```

```python
# Initial unemployment data import
# We using FRED :D

key = "aa9cd57aae80525dc171dbc517b39546"

url = "https://api.stlouisfed.org/fred/series/observations"

params = {
    "series_id": "NYICLAIMS",
    "api_key": key,
    "file_type": "json",
    "frequency": "w"
    #observation_start: (when we want to change the start time)
}

fred_initial_response = requests.get(url, params = params)
fred_initial_data = fred_initial_response.json()
```

```python
# Creating Initial FRED Dataframe

fred_initial_df = pd.DataFrame(fred_initial_data["observations"])
fred_initial_df
```

```python
# Continued unemployment data

params = {
    "series_id": "NYCCLAIMS",
    "api_key": key,
    "file_type": "json",
    "frequency": "w"
    #observation_start: (when we want to change the start time)
}
fred_continued_response = requests.get(url, params = params)
fred_continued_data = fred_continued_response.json()
```
