```python
# Import NYC Eviction dataset from NYC Open Data via json

# https://data.cityofnewyork.us/City-Government/Evictions/6z8x-wfk4/about_data
eviction_data = pd.DataFrame()
url = "https://data.cityofnewyork.us/resource/6z8x-wfk4.json"
limit = 1000
offset = 0
loop = True
while loop == True:
    params = {"$limit": limit, "$offset": offset}
    r = requests.get(url, params=params)
    data = r.json()
    clean_pop = pd.json_normalize(data)
    eviction_data = pd.concat([eviction_data, clean_pop])

    if len(data) < limit:
        loop = False

    offset += limit

# View eviction data

eviction_data
```

```python
# Import NYC Eviction dataset from NYC Open Data via json

# https://data.cityofnewyork.us/City-Government/Evictions/6z8x-wfk4/about_data
health_data = pd.DataFrame()
url = "https://data.cityofnewyork.us/resource/csut-3wpr.json"
limit = 1000
offset = 0
loop = True
while loop == True:
    params = {"$limit": limit, "$offset": offset}
    r = requests.get(url, params=params)
    data = r.json()
    clean_pop = pd.json_normalize(data)
    health_data = pd.concat([health_data, clean_pop])

    if len(data) < limit:
        loop = False

    offset += limit

# View eviction data

health_data
```
